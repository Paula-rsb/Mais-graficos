```python
import matplotlib.pyplot as plt
labels = 'Ciências Exatas', 'Ciências da Saúde', 'Ciências exatas e da Terra', 'Ciências Humanas', 'Ciências Sociais Aplicadas', 'Engenharias'
sizes = [410, 860, 3060, 1130, 3560, 990]
colors = ['blue','red','yellow','green','purple','cyan']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.0f%%', startangle=90)
plt.title (' Áreas de Conhecimento dos Grupos de Pesquisa em Tecnologia da Informação - Arredondado')
plt.axis('equal')
#fonte: http://www.scielo.br/j/cpa/a/TmJbqvbcBzdc9hkFGmjgHZK/?lang=pt# - gráfico 9
plt.show()
```
